```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('admission_data.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,0:-1]
Y = df.iloc[:,-1]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

'''
can also be used, but for better convergence, MinMaxScaler is highly recommended
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
'''
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
X_train.shape
'''
we have 400 datapoints having 7 features each, which are the deciding factor on whether a person will be receiving graduate
admission or not.
'''
```

```python
model = Sequential()
'''
7 features, so 7 input nodes
'''
model.add(Dense(7,activation='relu',input_dim=7))
model.add(Dense(7,activation='relu'))#7 nodes in the hidden layer
model.add(Dense(1,activation='linear'))#1 node in the o/p layer

'''
if you notice, we have the actual o/p y in the continuous value form. so obviously we will be using linear activation function
in the o/p layer.
'''
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
history = model.fit(X_train_trf,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred = model.predict(X_test_trf)
```

```python
'''from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
THE ABOVE WON'T work as it is for classification problem. we learned that for regression problems, we have r square (r^2) as our accuracy
calculator.
'''
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
